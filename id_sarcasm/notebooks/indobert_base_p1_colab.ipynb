## Visualisasi Training — Matplotlib

Plot train loss, eval loss, dan eval F1 per epoch untuk semua eksperimen.  
Bisa dijalankan di tengah training (menampilkan eksperimen yang sudah selesai) atau setelah semua eksperimen selesai.

```python
import json, os
import matplotlib.pyplot as plt

output_dirs = [
    ("Reddit — Baseline",  "outputs/indobert-base-p1-reddit-indonesia-sarcastic-augment"),
    ("Reddit — Weighted",  "outputs/indobert-base-p1-reddit-indonesia-sarcastic-weighted"),
    ("Twitter — Baseline", "outputs/indobert-base-p1-twitter-indonesia-sarcastic-augment"),
    ("Twitter — Weighted", "outputs/indobert-base-p1-twitter-indonesia-sarcastic-weighted"),
]

fig, axes = plt.subplots(len(output_dirs), 3, figsize=(18, 4 * len(output_dirs)))
fig.suptitle("IndoBERT Base-P1 — Training Curves", fontsize=15, fontweight="bold", y=1.01)

for row, (label, d) in enumerate(output_dirs):
    state_file = os.path.join(d, "trainer_state.json")
    if not os.path.exists(state_file):
        for col in range(3):
            axes[row, col].text(0.5, 0.5, f"{label}\n(belum selesai)",
                                ha="center", va="center", transform=axes[row, col].transAxes, fontsize=11)
        continue

    with open(state_file) as f:
        state = json.load(f)

    log_history = state["log_history"]
    best_epoch  = state.get("best_epoch")

    train_epochs, train_loss = [], []
    eval_epochs, eval_loss, eval_f1 = [], [], []

    for entry in log_history:
        if "loss" in entry and "eval_loss" not in entry:
            train_epochs.append(entry["epoch"])
            train_loss.append(entry["loss"])
        if "eval_loss" in entry:
            eval_epochs.append(entry["epoch"])
            eval_loss.append(entry["eval_loss"])
            eval_f1.append(entry.get("eval_f1"))

    # --- Train Loss ---
    ax = axes[row, 0]
    ax.plot(train_epochs, train_loss, color="steelblue", linewidth=1.5)
    if best_epoch:
        ax.axvline(x=best_epoch, color="red", linestyle="--", alpha=0.7, label=f"best epoch {best_epoch:.0f}")
        ax.legend(fontsize=8)
    ax.set_title(f"{label} — Train Loss", fontsize=10)
    ax.set_xlabel("Epoch"); ax.set_ylabel("Loss"); ax.grid(True, alpha=0.3)

    # --- Eval Loss ---
    ax = axes[row, 1]
    ax.plot(eval_epochs, eval_loss, color="darkorange", linewidth=1.5, marker="o", markersize=3)
    if best_epoch:
        ax.axvline(x=best_epoch, color="red", linestyle="--", alpha=0.7, label=f"best epoch {best_epoch:.0f}")
        ax.legend(fontsize=8)
    ax.set_title(f"{label} — Eval Loss", fontsize=10)
    ax.set_xlabel("Epoch"); ax.set_ylabel("Loss"); ax.grid(True, alpha=0.3)

    # --- Eval F1 ---
    ax = axes[row, 2]
    valid_f1 = [v for v in eval_f1 if v is not None]
    if valid_f1:
        ax.plot(eval_epochs[:len(valid_f1)], valid_f1, color="seagreen", linewidth=1.5, marker="o", markersize=3)
        best_f1 = max(valid_f1)
        ax.axhline(y=best_f1, color="red", linestyle="--", alpha=0.7, label=f"best F1 {best_f1:.4f}")
        ax.legend(fontsize=8)
    ax.set_title(f"{label} — Eval F1", fontsize=10)
    ax.set_xlabel("Epoch"); ax.set_ylabel("F1 Score"); ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("training_curves.png", dpi=150, bbox_inches="tight")
plt.show()
print("Plot disimpan ke training_curves.png")
```

```python
import json, os

output_dirs = [
    "outputs/indobert-base-p1-reddit-indonesia-sarcastic-augment",
    "outputs/indobert-base-p1-reddit-indonesia-sarcastic-augment-weighted",
    "outputs/indobert-base-p1-twitter-indonesia-sarcastic-augment",
    "outputs/indobert-base-p1-twitter-indonesia-sarcastic-augment-weighted",
]

for d in output_dirs:
    eval_file = os.path.join(d, "eval_results.json")
    if os.path.exists(eval_file):
        with open(eval_file) as f:
            results = json.load(f)
        print(f"\n{d.split('/')[-1]}")
        for k, v in results.items():
            print(f"  {k}: {v}")
    else:
        print(f"\n{d.split('/')[-1]}: eval_results.json belum ada")
```
